```python
import random

A=[]
m=int(input("satır:"))
n=int(input("sütun"))

for i in range(m):
    A+=[[]]
    for j in range(n):
        A[i]+=[random.randrange(1,10)]
for i in A:
    print(i)
print("************")


for p in range(m*n):
    for i in range(m):
        for j in range(n):
            if (j!=n-1) and (A[i][j]>A[i][j+1]):
                c=A[i][j+1]
                A[i][j+1]=A[i][j]
                A[i][j]=c
            
            if (i!=m-1) and (j==n-1) and (A[i][j]>A[i+1][0]):
                s=A[i+1][0]
                A[i+1][0]=A[i][j]
                A[i][j]=s

                
                
                
for i in A:
    print(i)
```
